```python
import re
import pandas as pd
import numpy as np
from tqdm import tqdm 
from copy import deepcopy
```

```python
def replace_str_lst(df):
    new_df = deepcopy(df).drop(columns=[f"x{i}" for i in range(8, 43)])

    for i in tqdm(range(8, 43)):
    # for i in tqdm(range(8, 9)):
        new_col = [str2lst(value) for value in df[f"x{i}"]]
        new_df.insert(i+1, f"x{i}", new_col)

    return new_df
```

```python
def str2lst(str_in):
    str_1 = str_in.replace('[', '')
    str_2 = str_1.replace(']', '')
    lst = re.split('\n | |, ', str_2)
    lst = [np.nan if i=='nan' else float(i) for i in lst]
    return lst
```

```python
new_df = replace_str_lst(df_train01)
```
