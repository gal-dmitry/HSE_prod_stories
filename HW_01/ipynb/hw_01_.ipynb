```python
# объединим датасеты, чтобы удобнее было вносить изменения

train01['part'] = 'train'
train02['part'] = 'val'
new_df = pd.concat([train01, train02]).reset_index(drop=True) 
```

```python
# удалим признаки, имеющие высокий процент nan

nans = '[nan, nan, nan, nan, nan, nan]'
xi = []

for i in range(10, 43):
    ratio = new_df[f"x{i}"][new_df[f"x{i}"] == nans].shape[0] / new_df[f"x{i}"].shape[0]
    pair = (i, ratio)
    xi.append(pair)
```

```python
xi
```

```python
xi_to_del = []

for (i, ratio) in xi:
    if ratio >= 0.7:
        ind = f"x{i}"
        xi_to_del.append(ind)
```

```python
xi_to_del
```

```python
new_df = new_df.drop(columns=xi_to_del)
```

```python
new_df.head()
```

```python
# конвертируем временные ряды для 6 месяцев в отдельные признаки по каждому месяцу
```

```python
series_lst = new_df.columns.tolist()[11:-1]
```

```python
series_lst
```

```python
for col in tqdm(series_lst):
    
    values = np.array([list(map(float, i[1:-1].split(','))) for i in new_df[col]])
    values[np.isnan(values)] = 0. #nan -> 0.
    
    col_names = [f"{col}_{i}" for i in range(1, 7)]
    new_df[col_names] = values
    new_df.drop(col, axis=1, inplace=True)
```
